```python
# load_predict.py

import numpy as np
import joblib
from tensorflow.keras.models import load_model
import warnings
warnings.filterwarnings('ignore')
# Load the trained model
model = load_model('ann_model.h5')

# Load the scaler
scaler = joblib.load('input_scaler.pkl')
```

```python
new_input = np.array([[50.0,1.0,18.0,1.0,5.5,2.0,1.0,5.0,2.0,8.0,3.0,2.0,1.0,12.0,2.0,2.0,0.0,5.0,150.0,240.0,130.0]])  # shape: (1, n_features)

# Scale the input
new_input_scaled = scaler.transform(new_input)

# Make prediction
prediction = model.predict(new_input_scaled)
predicted_class = np.argmax(prediction, axis=1)[0]
cls=['High','Low','Medium']
print(f"Predicted Class: {cls[predicted_class]} \n index {predicted_class}")
```
